```python
import numpy as np
from numpy.random import default_rng
import pandas as pd
import scipy.optimize as opt
import matplotlib.pyplot as plt
import seaborn as sns
from typing import NamedTuple, Dict
from sklearn import linear_model
```

```python
def calc_Y(A: np.ndarray, U: np.ndarray, alpha: float, beta: float) -> np.ndarray:
    """
    Y = cos(alpha + beta*A + U)
    """
    return np.cos(alpha + beta*A + U)

def calc_A(Z: np.ndarray, U: np.ndarray, e: np.ndarray, pi: float, delta: float) -> np.ndarray:
    """
    A = pi*Z + delta*U + e
    """
    return pi*Z + delta*U + e
        

def generatate_data(n_sample: int, params,  seed=1001, Z_noise: float = np.sqrt(3), e_Noise: float = 1):
    rng = default_rng(seed=seed)
    alpha, beta, pi, delta = params
    
    U = rng.uniform(-1, 2, n_sample)
    e = rng.normal(0, e_Noise, n_sample)
    Z = rng.normal(0, Z_noise, n_sample)
    A = calc_A(Z, U, e, pi, delta) 
    Y = calc_Y(A, U, alpha, beta)
    return U, e, Z, A, Y

def cal_true_effect(a: float, params, n_MC: int = 10000):
    """
    To estimate true effect
    n_MC affects MC error, seed is fixed!
    """
    rng = default_rng(seed=42)
    alpha, beta, pi, delta = params
    
    U = rng.uniform(-1, 2, n_MC)
    Y = calc_Y(a, U, alpha, beta)
    return np.mean(Y)


def generate_struc_data(params, a_start: float, a_end: float, n_steps: int = 100, n_MC: int = 10000):
    """
    from a_start to a_end, calc Y effect
    """
    a_steps = np.linspace(a_start, a_end, n_steps)
    y_struc = np.array([cal_true_effect(a, params, n_MC) for a in a_steps])
    return a_steps, y_struc
```

```python
def cal_Y_cond_Z(z: float, params, e_Noise: float = 1, n_MC: int = 10000):
    """
    E[Y|Z=z]
    """
    rng = default_rng(seed=42)
    alpha, beta, pi, delta = params
    
    U = rng.uniform(-1, 2, n_MC)
    e = rng.normal(0, e_Noise, n_MC)
    a = calc_A(z, U, e, pi, delta)
    Y = calc_Y(a, U, alpha, beta)
    return np.mean(Y)

def generate_cond_data(params, z_start: float, z_end: float, n_steps: int = 100, n_MC: int = 10000):
    z_steps = np.linspace(z_start, z_end, n_steps)
    y_cond = np.array([cal_Y_cond_Z(z, params, n_MC) for z in z_steps])
    return z_steps, y_cond
```

```python
def cal_h(z: float, params, e_Noise: float = 1, n_MC: int = 10000):
    """
    E[h(A)|Z=z] ?= E[cos(alpha + beta * a)|Z=z]
    """
    rng = default_rng(seed=42)
    alpha, beta, pi, delta = params
    
    U = rng.uniform(-1, 2, n_MC)
    e = rng.normal(0, e_Noise, n_MC)
    a = calc_A(z, U, e, pi, delta)
    
    h_a = np.cos(alpha + beta*a)
    return np.mean(h_a)

def generate_h_data(params, z_start: float, z_end: float, n_steps: int = 100, n_MC: int = 10000):
    z_steps = np.linspace(z_start, z_end, n_steps)
    h_a_cond = np.array([cal_h(z, params, n_MC) for z in z_steps])
    return z_steps, h_a_cond
```

### Weird sample behavior on n_MC! Not depending on seed! What is going on there?

```python
n_sample = 10000
random_seed = 1001
alpha_true = 1
beta_true = 2
pi_true = 3
delta_true = 1 
e_Noise = 1
Z_noise = np.sqrt(3) # more signal
params_true = np.array([alpha_true, beta_true, pi_true, delta_true])
n_MC = 10000


# generate Data
U_data, e_data, Z_data, A_data, Y_data = generatate_data(n_sample, params_true,  seed=random_seed, 
                                                         Z_noise = Z_noise, e_Noise = e_Noise)

# generate structural Data (True Effect)
a_steps, y_struc = generate_struc_data(params_true, a_start=-10, a_end=10, n_steps = 1000, n_MC = n_MC)

# generate E[Y|Z=z] 
z_steps, y_cond = generate_cond_data(params_true, z_start=-5, z_end=5, n_steps = 1000, n_MC = n_MC)

# generate E[Y|Z=z] approx
# z_steps_approx, y_cond_approx = generate_cond_approx_data(params_true, z_start=-5, z_end=5, n_steps = 1000, n_MC = n_MC)

# generate E[h(A)|Z=z]
z_steps_h_a_cond, h_a_cond = generate_h_data(params_true, z_start=-5, z_end=5, n_steps = 1000, n_MC = n_MC)
# generate E[h(A)|Z=z]1
z_steps_h_a_cond1, h_a_cond1 = generate_h_data(params_true, z_start=-5, z_end=5, n_steps = 1000, n_MC = 1000)
# generate E[h(A)|Z=z]2
z_steps_h_a_cond2, h_a_cond2 = generate_h_data(params_true, z_start=-5, z_end=5, n_steps = 1000, n_MC = 1000000)

# Plot Results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(x=A_data, y=Y_data, label='Observed', ax=ax1)
ax1.plot(a_steps, y_struc, color='orange', marker='o', label="True Effect", markersize=5)
ax1.set_title('Treatment / Outcome')
ax1.set_xlabel('A')
ax1.set_ylabel('Y')
ax1.legend()

sns.scatterplot(x=Z_data, y=Y_data, label='Observed', ax=ax2)
ax2.plot(z_steps, y_cond, color='orange', marker='o', label="E[Y|Z]", markersize=3)
ax2.plot(z_steps_h_a_cond, h_a_cond, color='red', marker='o', label=f"E[h(A)|Z], MC: {n_MC}", markersize=3)
ax2.plot(z_steps_h_a_cond1, h_a_cond1, color='green', marker='o', label="E[h(A)|Z], MC:1000", markersize=3)
ax2.plot(z_steps_h_a_cond2, h_a_cond2, color='gray', marker='o', label="E[h(A)|Z], MC: 100000", markersize=3)
# ax2.plot(z_steps_approx, y_cond_approx, color='violet', marker='o', label="E[Y|Z] (Approx)", markersize=5)
ax2.set_title('Instrument / Outcome')
ax2.set_xlabel('Z')
ax2.set_ylabel('Y')
ax2.set_ylim([-0.025, 0.025])
ax2.set_ylabel('Y')
ax2.legend()

fig.suptitle('Visualization')
plt.tight_layout()
plt.show()
```
